**Question 3 :Write a Python script to create a data frame for the following table**

*  Create a bar plot of the sales volume.
*  Create a pie chart of item sales.
*  Create an area plot of item sales.
*  Create a stacked bar plot of item sales



```python
d = {'Name': ['Ahmed','Omar','Ali','Ziad','Salwa','Lila'], 'Mobile_Sales': [2540,1370,1320,2000,2100,2150], 'TV_Sales':[2200,1900,2150,1850,1770,2000]}
df = pd.DataFrame(data= d)
df
```

```python
a = df['Mobile_Sales'].sum()
b = df['TV_Sales'].sum()
z = {'Name' : ['Mobile_Sales','TV_Sales'],'Sum' : [a ,b]}
df2 = pd.DataFrame(data=z,index=['Mobile_Sales', 'TV_Sales'])
df2
```

```python
df2.plot.bar(x='Name',figsize = (20,10)).legend(bbox_to_anchor=(1.1,1))
```

```python
df2.plot.pie(x='Name', y ='Sum',figsize = (20,10))
```

```python
df.plot.area(x='Name',figsize=(20,10)).legend(bbox_to_anchor=(1.3,1))
```

```python
df.plot.bar(x='Name',stacked=True,figsize=(20,10)).legend(bbox_to_anchor=(1.1,1))
```
