##**Import Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plot style
plt.style.use('seaborn-whitegrid')
#Provide values on x and y
X = [590,540,740,130,810,300,320,230,470,620,770,250]
Y = [32,36,39,52,61,72,77,75,68,57,48,48]
```

```python
#Use scatter plot
plt.scatter(X,Y)
#Determine the limitation on X and Y in plot
plt.xlim(0,1000)
plt.ylim(0,100)
```

```python
#Scatter plot color
plt.scatter(X,Y, s=60 , c='green',marker='+')
plt.xlim(0,800)
plt.ylim(0,90)
#Add title
plt.title('Relationship Between Temperature and Iced Coffee Sales')
plt.xlabel('Sold Coffee')
plt.ylabel('Temperature in Fahrenheit')

#show plot
plt.show()
```

```python
X = [590,540,740,130,810,300,320,230,470,620,770,250]
Y = [32,36,39,52,61,72,77,75,68,57,48,48]
plt.scatter(X,Y, s=90 , c='green',marker='+')
plt.xlim(0,900)
plt.ylim(0,90)

#Add title
plt.title('Relationship Between Temperature and Iced Coffee Sales')
plt.xlabel('Sold Coffee')
plt.ylabel('Temperature in Fahrenheit')

#Add the trend line
z = np.polyfit(X, Y, 1)
p = np.poly1d(z)
plt.plot(X,p(X),"r-")

#show plot
plt.show()
```

```python
#Create empty figure
fig = plt.figure()
#Create ax plot
ax = plt.axes()
#Generate x value using linespace
x = np.linspace(0,5,1000)
```

#**Lineplot**

```python
#Plot Y = sin(x) , Y = cos(x)
ax.plot(x,np.sin(x));
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

#Set the x and y axis range
plt.xlim(0,11)
plt.ylim(-2,2)
plt.axis('tight')
#Add title 
plt.title('Plotting data using sin and cos')
#Show plot
plt.show()
```

```python
x = np.linspace(0,50,1000)
#Plot Y = sin(x) , Y = cos(x)
ax.plot(x,np.sin(x));
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

#Set the x and y axis range
plt.xlim(0,11)
plt.ylim(-2,2)
plt.axis('auto')
#Add title 
plt.title('Plotting data using sin and cos')
#Show plot
plt.show()
```
