```python
import matplotlib.pyplot as plt
```

```python
plt.plot(phd,label = 'Ph.D.')
plt.plot(service,label='service')
plt.xlabel('Ph.D./service')
plt.ylabel('Frequency')
plt.title('Ph.D./service\nDistribution')
plt.legend()
plt.show()
```

```python
plt.plot(phd,service,'bo',label = "Ph.D. Vs services",lw=10)
plt.grid()
plt.legend()
plt.xlabel('Ph.D')
plt.ylabel('service')
plt.title('Ph.D./salary\nDistribution')
plt.yscale('log')
```

```python
x = np.arange(10)
width = 0.35
plt.bar(x - width/2,phd[:10],width,label='phd',color='r')
plt.bar(x + width/2,service[:10],width,label="service",color='c')
plt.xlabel('phd')
plt.ylabel('Total number')
plt.title('Ph.D. Vs Service \n 2018')
plt.legend()
plt.show()
```

```python
plt.hist(service,bins=30,rwidth=0.8,color='green',label='service')
plt.hist(phd,bins=30,rwidth=0.8,color='red',label='phd')
plt.xlabel('Services/phd')
plt.ylabel('Distribution')
plt.title('Service/phd\n2018')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.scatter(phd,salary,label='Salary/phd',color='g',marker='+',s=80)
plt.xlabel('phd')
plt.ylabel('salary')
plt.title('phd/salary\nSpring 2018')
plt.legend()
plt.show()
```

```python
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
plt.plot([],[],color='m',label='Sleeping')
plt.plot([],[],color='c',label='Eating')
plt.plot([],[],color='r',label='Working')
plt.plot([],[],color='k',label='Playing')
plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])
plt.xlabel('days')
plt.ylabel('Activities')
plt.title('Persons Weekly Spent Time per activities\nSpring 2018')
plt.legend()
plt.show()
```
